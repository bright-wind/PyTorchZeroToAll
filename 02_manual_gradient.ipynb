```python
x_data = [1.0, 2.0, 3.0]
y_data = [2.0, 4.0, 6.0]

w = 1.0                                       # for gradient descent, make a random guess for the weight w

# our model forward pass                      # this is our model
def forward(x):
    return x * w


# Loss function                               # here is the error or loss function prediction - actual
def loss(x, y):
    y_pred = forward(x)
    return (y_pred - y) * (y_pred - y)


# compute gradient                            # this is the gradient (derivative) for the foward pass model
def gradient(x, y):  # d_loss/d_w
    return 2 * x * (x * w - y)

# Before training
print("\n predict (before training) w = ", w ," - Study 4 hours to receive", round(forward(4)))

# Training loop
print ('\n TRAINING LOOP')
for epoch in range(25):
    for x_val, y_val in zip(x_data, y_data):      #  iterate through each input feature and its response
        grad = gradient(x_val, y_val)             # calculate the gradient / derivative at this w
        w = w - 0.01 * grad                       # update the weight with a small fraction (learning rate) of the gradient
        print("\tgrad: ", x_val, y_val, round(grad, 2))
        l = loss(x_val, y_val)

    print("progress:", epoch, "w=", round(w, 2), "loss=", round(l, 2))

# After training
print("\n predict (after training) w = ", round(w),  " - Study 4 hours to receive", round(forward(4)))
```
